```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
nodos = 50
x = []
y = []
tipos = []
with open("instances/TTRP_03.txt",'r') as file:
    lines = file.readlines()
    for line in lines[1:]:
        _,xPos,yPos,_,tipo = line.strip().split()
        tipos.append('o' if int(tipo) == 0 else 's')
        x.append(float(xPos))
        y.append(float(yPos))
clientes = pd.DataFrame(data={'x':x,'y':y,'tipo':tipos})

def plot_clients():
    for i in range(nodos):
        plt.scatter(x[i],y[i],marker=tipos[i],color='r')
```

```python
# Leer rutas
lines = []
with open("out_initial.txt",'r') as file:
    lines = file.readlines()
colors = ['r','g','b','k','m']
count = 0
for line in lines:
    ruta = list(map(int,line.strip().split(" ")))
    color = colors[count]
    for i in range(len(ruta)-1):
        l = '-'
        plt.plot([x[ruta[i]],x[ruta[i+1]]],[y[ruta[i]],y[ruta[i+1]]],color+l)
    count+=1

plot_clients()
plt.scatter(x[0],y[0],color='b')
plt.show()
```

```python
# Leer rutas
lines = []
with open("out_final.txt",'r') as file:
    lines = file.readlines()
colors = ['r','g','b','k','m']
count = 0
for line in lines:
    ruta = list(map(int,line.strip().split(" ")))
    color = colors[count]
    for i in range(len(ruta)-1):
        l = '-' if tipos[ruta[i]] == 'o' and tipos[ruta[i+1]] == 'o' else '--'
        plt.plot([x[ruta[i]],x[ruta[i+1]]],[y[ruta[i]],y[ruta[i+1]]],color+l)
    count+=1

plot_clients()
plt.scatter(x[0],y[0],color='b')
plt.show()
```
